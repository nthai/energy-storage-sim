```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 7]
plt.rcParams['figure.dpi'] = 600

df0 = pd.read_csv('tmp_log1.log') # 2 Li-ion, 3 Flywheel, 0 Supercapacitor
df1 = pd.read_csv('tmp_log2.log') # 2 Li-ion, 3 Flywheel, 2 Supercapacitor

df0['timestamp'] = pd.to_datetime(df0['timestamp'])
df1['timestamp'] = pd.to_datetime(df1['timestamp'])
```

```python
start = df0.iloc[0]['timestamp']
end = df0.iloc[-10]['timestamp']
plt.xlim(start, end)

plt.plot(df0['timestamp'], df0['pnet'], linewidth=.5)
plt.plot(df0['timestamp'], df0['upper'], linewidth=1.5)
plt.plot(df0['timestamp'], df0['lower'], linewidth=1.5)
```

```python
start = df0.iloc[0]['timestamp']
end = df0.iloc[500]['timestamp']
plt.xlim(start, end)
plt.ylim(0, 80)

plt.plot(df0['timestamp'], df0['pnet'], linewidth=.5, color='black')
plt.plot(df0['timestamp'], df0['pbought'], linewidth=.5, color='green')
plt.plot(df1['timestamp'], df1['pbought'], linewidth=.5, color='orange')
plt.plot(df0['timestamp'], df0['upper'], linewidth=.5, color='red')
plt.plot(df1['timestamp'], df1['lower'], linewidth=.5, color='blue')
plt.plot(df0['timestamp'], df0['soc'], linewidth=.5, color='green')
plt.plot(df1['timestamp'], df1['soc'], linewidth=.5, color='orange')
```
